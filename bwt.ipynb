```python
def approximatePatternMatching( p, s, d ):
    return [i for i in range(len(s)-len(p)+1) if hamming(p,s[i:i+len(p)]) <= d]

def hamming( s, t ):
    return sum([s[i] != t[i] for i in range(len(s))])
```

```python
import sys
from functools import partial

def radix_sort(values, digit, shift=0):
    if len(values) <= 1:
        for v in values:
            yield v
        return
    buckets = {}
    for v in values:
        buckets.setdefault(digit(v, shift), []).append(v)
    for k in sorted(buckets.keys()):
        for r in radix_sort(buckets[k], digit, shift + 1):
            yield r


def l(s, value, shift):
    if (value + shift > len(s) - 1):
        print('shit')
    return s[(value + shift) % len(s)]


def bwt(s):
    return ''.join(s[i - 1] for i in radix_sort(range(len(s)), partial(l, s)))


s = ''
with open('genome.fa', 'r') as f:
    for line in f:
        if line.startswith(">"): continue
        s += line.strip()
s += '$'
f.close()
orig_stdout = sys.stdout
f = open('bwt.txt', 'w')
sys.stdout = f
print(bwt(s))
sys.stdout = orig_stdout
f.close()
```
